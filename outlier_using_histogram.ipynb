        ### Z_Score ###

```python


def detect_outliers(data):
    threshold = 2 # 2rd standard deviation
    mean = np.mean(data)
    std = np.std(data)
    outliers = []  # Define outliers inside function to avoid global variable issues

    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)  # Append the outlier value

    return outliers
```

```python
detect_outliers(ds)
```
