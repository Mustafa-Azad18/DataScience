```python

from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python

import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
     

reg = LinearRegression()
reg.fit(X,y)
     
```

```python

reg.coef_
     
# array([78.35063668])

```

```python
reg.intercept_
```

```python

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
     
```

```python

# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
y_pred = ((78.35 * X) + 100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
plt.scatter(X, y)

# OLS line
plt.plot(X, reg.predict(X), color='red', label='OLS')

# Line with b = 0 (intercept = 100)
plt.plot(X, y_pred, color='#00a65a', label='b = 0')

# Force the x-axis to include 0
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)  # show vertical line at x=0
plt.ylim(0, 200)   # adjust y-axis so intercept is visible
plt.xlim(-1, 1)    # extend x-axis so 0 is inside the range

plt.legend()
plt.show()
```

```python

b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
