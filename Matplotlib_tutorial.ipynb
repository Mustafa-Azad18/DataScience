# Line plot #
#### This is used to visualize data points connected by a continuous line ####

To Track chnages in a variable of over time
To compare 
To detect patterns or anomalies in data



```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
product_data ={
    "Month":['jan','Feb','Mar','Apr','May','June','July'],
    "Stock_value": [2000,2058,2100,2245,2300,2350,2500]
}
```

```python
df = pd.DataFrame(product_data)
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Month'],df['Stock_value'],marker='o',linestyle='-',color='r')
plt.title("Stock value over time")
plt.xlabel("Month")
plt.ylabel("Stock_value")
plt.grid()
plt.show()
```

# bar plot
## A bar plot represents a categorical data using rectengular  bars , where the length of each bar is proportunal to the value it represents
## Situations where we use bar plot areabs 
 To compare different categories or groups, 
 To highlight differences in various datapoint,
 To analyze frequency distributions  


```python
plt.figure(figsize=(8,5))
plt.bar(df['Month'],df['Stock_value'],color='red',edgecolor='black')
plt.title("Stock value in different month")
plt.xlabel("Month")
plt.ylabel("Stock_value")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

A histogram is a graphycal representation of the data distribution(Numarical data) using bin , It shows the frequency of different data ranges, the situations when we use a histogram-
To understand how data is distributed


```python
plt.figure(figsize=(8,5))
plt.hist(df['Stock_value'], bins=4, color='orange', edgecolor='black')
plt.title("Distribution of stock values")
plt.xlabel("Stock_value")
plt.ylabel("Frequency")
plt.show()
```

# Scatterplot
### It displays datapoints to show the relatinally between the numarical vlaues
## Situations where we use scatterplot
### 1.To detect co-relations
### 2. To identify clusters and outliers in the dataset

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Month'],df['Stock_value'],color='purple')
plt.title("Stock value vs month ")
plt.xlabel("Stock_value")
plt.ylabel("Month")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid()
plt.show()
```

# Pie chart
### Pie chart is a circular statistical graphic, dividev into slices to illustrate the proportion of different categories, situations where we use pie chart are--
### 1. To show the different categories proportions
### 2. To compare the parts of whole 
### 3. To represent percentage distribuiton

```python
plt.figure(figsize=(8,6))
plt.pie(df["Stock_value"], labels=df["Month"],autopct='%1.2f%%',startangle=140, colors=plt.cm.Paired.colors)
plt.title("Monthly Stock Values Distribution")
plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))

# First subplot (1 row, 2 columns, 1st plot)
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue', linestyle='-', label='sin(x)')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

# Second subplot (1 row, 2 columns, 2nd plot)
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red', linestyle='--', label='cos(x)')
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')      # Other examples: 'seaborn', 'bmh', 'dark_background'
```

```python
plt.plot(x, y1, color='navy', linestyle='--', linewidth=2, marker='o', markersize=5)
plt.title('Styled Line Plot', fontsize=14, fontweight='bold', color='darkred')
```
