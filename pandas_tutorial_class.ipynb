```python
num=pd.Series([10,20,30,40,50])
print("Original series:\n",num)
print("Adding 5 in the Original series:\n",num+5)
print("Multiply 2 Original series:\n",num*2)
print("Square of Original series:\n",num**0.5)
```

```python
data_with_NaN=pd.Series([10,np.nan,30,np.nan,50],index=['A','B','C','D','E'])
```

```python
data_with_NaN.notnull()
```

```python
print("Original series with missing values:", data_with_NaN)
print("Check for Nan values :", data_with_NaN.isnull())
```

```python
filled_series=data_with_NaN.fillna(0)
```

```python
print("Filling missing values with 0:",filled_series)
```

```python
dropped_series= data_with_NaN.dropna()
print("Droping missing values:", dropped_series)
```
